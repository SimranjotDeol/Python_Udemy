```python
# DATA
df = pd.DataFrame(np.random.randn(100,4),columns = ['A','B','C','D'])
```

```python
df.head()
```

```python
df2 = pd.DataFrame({'Category':['A','B','C'], 'Values': [32,43,50]})
```

```python
df2
```

```python
df.iplot()
```

```python
df.iplot(kind = 'scatter', x = 'A', y = 'B') #Plotly will try to connect all the points by itself
```

```python
df.iplot(kind = 'scatter', x = 'A', y = 'B', mode = 'markers',size = 10) #Markers input disables the connecting lines
```

```python
df2.iplot(kind = 'bar', x = 'Category', y ='Values')
```

```python
df.sum().iplot(kind = 'bar') #Plotting df by summing accross the columns for each Category
```

```python
df.iplot(kind = 'box') #Can click on legend to hide certain boxes
```

```python
df3 = pd.DataFrame({'x':[1,2,3,4,5], 'y':[10,20,30,20,10],'z':[5,4,3,2,1]})
```

```python
df3
```

```python
df3.iplot(kind = 'surface',colorscale= 'rdylbu')
```

```python
df['A'].iplot(kind = 'hist',bins = 35)
```

```python
df.iplot(kind = 'hist')
```

```python
df[['A','B']].iplot(kind = 'spread')

# Useful for stock data
```

```python
df.iplot(kind = 'bubble', x='A', y='B',size='C')

#Useful for world GDP vs population
```
